```python
# consts
VALIDATION_SPLIT = 0.1
SEED=np.random.randint(1,1000)
SEED
```

```python
train_df = pd.read_csv(path_train)
test_df = pd.read_csv(path_test)
```

### converting date features to integers

```python
train_df["last_payment_int"] = train_df["last_payment"].map(helpers.convert_date_to_number_of_days_from_today)
train_df["end_last_loan_int"] = train_df["end_last_loan"].map(helpers.convert_date_to_number_of_days_from_today)

test_df["last_payment_int"] = test_df["last_payment"].map(helpers.convert_date_to_number_of_days_from_today)
test_df["end_last_loan_int"] = test_df["end_last_loan"].map(helpers.convert_date_to_number_of_days_from_today)
```

## simple baseline

```python
numerical_column_names_to_use = ["score_3",
                                "score_4",
                                "score_5",
                                "score_6",
                                "risk_rate",
                                "amount_borrowed",
                                "credit_limit",
                                "income",
                                "n_bankruptcies",
                                "n_defaulted_loans",
                                "last_payment_int",
                                "end_last_loan_int",
                                "n_issues"]

categorical_column_names_to_use = ["score_1",
                                   "score_2",
                                   "gender",
                                  "facebook_profile"]

normalized_train_df = helpers.get_normalized_df_simple(train_df,
                                                       numerical_column_names_to_use,
                                                       categorical_column_names_to_use)
normalized_test_df = helpers.get_normalized_df_simple(test_df,
                                                      numerical_column_names_to_use,
                                                      categorical_column_names_to_use,
                                                      test=True)
```

```python
train_data = normalized_train_df.values
test_data = normalized_test_df.values

train_data.shape,test_data.shape
```

```python
indices = np.arange(train_data.shape[0])
np.random.shuffle(indices)

train_data = train_data[indices]

X_train= train_data[:,1:]
Y_train = train_data[:,0]

X_test = test_data


X_train.shape,Y_train.shape, X_test.shape
```

```python
X_train.shape,Y_train.shape
```

```python
clf = LogisticRegression(random_state=SEED)
```

```python
% time
# using stratified shuffle split because the number of rows with default == True is much smaller
# than the number of rows with default == False so we want to keep the same proportions in all splits
cv = StratifiedShuffleSplit(n_splits=10, test_size=VALIDATION_SPLIT, random_state=SEED)

# estimating the log_loss score on the various splits
cross_val_score(clf, X_train, Y_train, cv=cv,scoring='neg_log_loss')
```

```python
% time

clf.fit(X_train,Y_train)
```

```python
# debugging
y_pred_proba = clf.predict_proba(X_test[0].reshape(1,-1))
y_pred_proba
```

## output

```python
ids = test_df["ids"].values

y_preds_binary = [clf.predict(x_test.reshape(1,-1)) for x_test in X_test]
y_preds_lst = ['True' if y_pred == 1 else 'False' for y_pred in y_preds_binary]

y_preds_np = np.array(y_preds_lst).reshape(-1,1)
```

```python
out_df = pd.DataFrame({'ids': ids, 'predictions': y_preds_lst})
```

```python
out_df.to_csv('predictions.csv',index=False)
```
