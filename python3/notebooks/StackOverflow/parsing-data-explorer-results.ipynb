```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from natsort import natsorted, ns
%matplotlib inline
```

```python
results = dict()

with open('/home/felipe/stackoverflow-data-explorer-count-by-tag-2016-2017.txt','r') as f:
    for line in f:
                
        tag,count,yearmonth =line.split()

        if yearmonth.strip()=='2017-10':
            continue
            # month has not ended 
        
        if results.get(tag.strip(),None) is None:
            results[tag.strip()] = [(yearmonth.strip(),int(count.strip()) )]
        else:
            results[tag.strip()] = results[tag.strip()] + [(yearmonth.strip(), int(count.strip()) )]
```

```python
periods = ['2016-01','2016-02','2016-03','2016-04','2016-05','2016-06',
           '2016-07','2016-08','2016-09','2016-10','2016-11','2016-12',
           '2017-01','2017-02','2017-03','2017-04','2017-05','2017-06',
           '2017-07','2017-08','2017-09']
```

```python
list(results.keys())
```

```python
for key in list(results.keys()):
    existing_values = [pair[0] for pair in results[key]]
    
    missing = set(periods) - set(existing_values)
    
    for period in missing:
        results[key] = results[key] + [(period,0)]   
```

```python
# merge pymc and pymc3
pymc3 = results['pymc3']
pymc3
```

```python

new_pairs = []

for (date,count) in results['pymc']:
    new_count = count + [pair[1] for pair in pymc3 if pair[0]==date][0]
    
    new_pairs.append((new_count,date))
    
new_pairs
```

```python
results.pop('pymc3',None)
results.pop('pymc',None)

results['pymc/pymc3'] = new_pairs
```

```python
for key in list(results.keys()):
    
    results[key] = natsorted(results[key],key=lambda pair: pair[0])
```

```python
plt.clf()

img = plt.gcf()

ax = plt.gca()
for key in list(results.keys()):
    x = periods
    y = [pair[1] for pair in results[key]]
    plt.plot(np.array(x),np.array(y),label=key)
    
plt.legend(loc='lower right')
plt.grid(True)

plt.yscale('symlog')
    
img.set_size_inches(20,12)
```
